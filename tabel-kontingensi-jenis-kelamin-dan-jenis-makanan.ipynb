```python
import numpy as np
from scipy.stats import chi2_contingency

# Tabel kontingensi:
# Baris = jenis makanan: [Olahan ketupat, Rendang, Mie ayam]
# Kolom = jenis kelamin: [Laki-laki, Perempuan]

observasi = np.array([
    [3, 5],  # Olahan ketupat
    [4, 3],  # Rendang
    [2, 3]   # Mie ayam
])

chi2, p_value, dof, expected = chi2_contingency(observasi)

print("HASIL UJI CHI-SQUARE")
print(f"Chi-square hitung : {chi2:.4f}")
print(f"p-value           : {p_value:.4f}")
print(f"Derajat bebas     : {dof}")
print("\nFrekuensi harapan (expected frequency):")
print(expected)
```

```python
alpha = 0.05

print("KEPUTUSAN")
if p_value < alpha:
    print("Karena p-value < 0.05, maka terdapat hubungan antara jenis kelamin dan jenis makanan.")
else:
    print("Karena p-value > 0.05, maka tidak terdapat hubungan yang signifikan antara jenis kelamin dan jenis makanan.")

# Tambahan: ukuran kekuatan hubungan dengan Cramer's V
n = observasi.sum()
r, c = observasi.shape
cramers_v = np.sqrt(chi2 / (n * min(r - 1, c - 1)))

print(f"\nCramer's V        : {cramers_v:.4f}")
```
